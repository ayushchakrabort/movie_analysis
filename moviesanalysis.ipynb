```python
import numpy as dp
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('movies.csv')
```

```python
data
```

```python
new_data1=data.melt(id_vars='Title',value_vars=['Netflix', 'Hulu','Prime Video', 'Disney+'],var_name='Platforms',value_name='Available')
# new_data1=new_data1[new_data1['Available']==1]
```

```python
data['Available_on']=data[['Netflix', 'Hulu','Prime Video', 'Disney+']].sum(axis=1)
```

```python
data['Genre_Count'] = data['Genres'].str.split(',').apply(len)
```

```python
data['Country Count'] = data['Country'].str.split(',').apply(len)
data['Language Count'] = data['Language'].str.split(',').apply(len)
```

```python
data.columns
```

```python
data
```

```python
new1_df=data[['Title','Genres']]
new1_df.loc[::,'Genres']=new1_df['Genres'].str.split(',')
df_genre=new1_df.explode('Genres')
df_genre
```

```python
new2_df=data[['Title','Language']]
new2_df['Language']=new2_df['Language'].str.split(',')
df_lang=new2_df.explode('Language')
df_lang
```

```python
new3_df=data[['Title','Directors']]
new3_df['Directors']=new3_df['Directors'].str.split(',')
df_dir=new3_df.explode('Directors')
df_dir
```

```python
new4_df=data[['Title','Country']]
new4_df['Country']=new4_df['Country'].str.split(',')
df_cou=new4_df.explode('Country')
df_cou
```

```python
new_data=df_cou.merge(df_dir,on='Title').merge(df_lang,on='Title').merge(df_genre,on='Title')
new_data
```

```python
new_data.describe(include='all')
```

```python
data.drop(['Netflix', 'Hulu','Prime Video', 'Disney+'],axis=1,inplace=True)
data.drop(['Country', 'Language','Genres','Directors'],axis=1,inplace=True)
movie_data=pd.merge(new_data,new_data1,on='Title')
moviedata=pd.merge(movie_data,data,on='Title')
moviedata
```

# Questions- 
Averge IMDb and rotten tomatoes rating in each platform??

```python
# Averge IMDb rating in each platform
ppp=moviedata[(moviedata['Available']==1) & (moviedata['IMDb']!=11)]
ppp.groupby('Platforms')['IMDb'].mean()
ppp.reset_index(inplace=True)
plt.figure(figsize=(17,6))
plt.subplot(2,2,2)
plt.title('Average IMDb rating')
obje=sns.barplot(ppp,x='Platforms',y='IMDb')
for i in obje.containers:
    obje.bar_label(i)

# Average Rotten tomato rating in each platform
qqq=moviedata[(moviedata['Available']==1) & (moviedata['Rotten Tomatoes']!=11)]
qqq.groupby('Platforms')['Rotten Tomatoes'].mean()
qqq.reset_index(inplace=True)
plt.subplot(2,2,1)
plt.title('Average RottenT rating')
obj=sns.barplot(qqq,x='Platforms',y='Rotten Tomatoes')
for i in obj.containers:
    obj.bar_label(i)


plt.tight_layout()
plt.show()
```

# Observation- 

"According to the graph, Prime Video has a lower average rating compared to the other platform."

# Question-
Comparing of high IMDb-rated movies to its low-rated movies, and what might be influencing this trend?

```python
# Number of movies available in each platforms
ppp=moviedata[(moviedata['Available']==1) & ((moviedata['IMDb']!=11) & (moviedata['Rotten Tomatoes']!=11))]
plt.figure(figsize=(30,8))
plt.subplot(2,5,1)
obj=sns.countplot(ppp,x='Platforms')
for i in obj.containers:
    obj.bar_label(i)

# Highest IMDb rating movies where they are streamed
p=moviedata[moviedata['IMDb']!=11]
# p['IMDb'].max()
a=p.nlargest(5000,'IMDb')
plt.figure(figsize=(30,8))
plt.subplot(2,5,2)
plt.title('Highest IMDb')
obj=sns.countplot(a,x='Platforms',hue='Available')
for i in obj.containers:
    obj.bar_label(i)

# Lowest IMDb rating movies where they are streamed
p=moviedata[moviedata['IMDb']!=11]
# p['IMDb'].min()
a=p.nsmallest(5000,'IMDb')
plt.subplot(2,5,3)
plt.title('Lowest IMDb')
objo=sns.countplot(a,x='Platforms',hue='Available')
for i in objo.containers:
    objo.bar_label(i)
plt.tight_layout
plt.show()
```

# Observation- 
1) Prime Video seems to be offering a good number of high IMDb-rated movies, but when you compare it to the low-rated ones, it appears to be releasing more low-rated movies overall.

2) Netflix and Hulu have released fewer low-rated movies, which seems fair, but there’s can be improvement.

3) Disney+ has made a strong impression with its high IMDb-rated movies, but it could add more movie, as the number of movies currently available is low.

# Question-
**How does the relationship between Rotten Tomatoes and IMDb ratings influence each other, and why do they follow a similar pattern in terms of movie ratings?**

```python
ppp=moviedata[(moviedata['Available']==1) & ((moviedata['IMDb']!=11) & (moviedata['Rotten Tomatoes']!=11))]
plt.figure(figsize=(30,8))
plt.subplot(2,5,1)
obj=sns.countplot(ppp,x='Platforms')
for i in obj.containers:
    obj.bar_label(i)


# For highest rotten rating and where they are streamed
r=moviedata[moviedata['Rotten Tomatoes']!=11]
r['Rotten Tomatoes'].max()
r_tomato=r.nlargest(5000,'IMDb')
plt.figure(figsize=(30,8))
plt.subplot(2,5,1)
plt.title('Highest RottenT')
obj=sns.countplot(r_tomato,x='Platforms',hue='Available')
for i in obj.containers:
    obj.bar_label(i)

# For lowest rotten rating and where they are streamed
rt=moviedata[moviedata['Rotten Tomatoes']!=11]
rt['Rotten Tomatoes'].min()
rt_tomato=rt.nsmallest(5000,'Rotten Tomatoes')
plt.subplot(2,5,2)
plt.title('Lowest RottenT')
objo=sns.countplot(rt_tomato,x='Platforms',hue='Available')
for i in objo.containers:
    objo.bar_label(i)
plt.tight_layout
```

# Observation-
1) It seems like rotten tomatoes and IMDb rating is having a quite strong relation with each other, as we analyze the IMDb ratting it also shows the same pattern.
2) Highest Rotten Tomatoes rated movies are been released in the Prime video but it has released more low rated movies in it Platform.
3) Netflix and Disney+ have streamed more High rated movies which is quite impressive but Disney+ is needed to increase it count.

# Question 
**How does the language in which a movie is released influence its IMDb rating**

```python
# Top 1000 IMDb rating movie count in which language they are released
pq=moviedata[(moviedata['Language']!='Unknown') & (moviedata['IMDb']!=11)]
filtered_lang=pq.nlargest(1000,'IMDb')
a=pd.DataFrame(filtered_lang.groupby('Language')['Title'].count())
a.reset_index(inplace=True)
plt.figure(figsize=(28,5))
plt.subplot(2,3,1)
obj=sns.barplot(a,x='Language',y='Title')
for i in obj.containers:
    obj.bar_label(i)

filtered_lan=pq.nsmallest(1000,'IMDb')
b=pd.DataFrame(filtered_lan.groupby('Language')['Title'].count())
b.reset_index(inplace=True)
plt.subplot(2,3,2)
obj=sns.barplot(b,x='Language',y='Title')
for i in obj.containers:
    obj.bar_label(i)
plt.tight_layout()

# Highest IMDb rating released on language
plt.figure(figsize=(28,5))
plt.subplot(2,3,1)
plt.title('IMDb rating released on language')
obj=sns.barplot(filtered_lang,x='Language',y='IMDb')
for i in obj.containers:
    obj.bar_label(i)

# Lowest IMDb rating released on language
plt.subplot(2,3,2)
plt.title('IMDb rating released on language')
obje=sns.barplot(filtered_lan,x='Language',y='IMDb')
for i in obje.containers:
    obje.bar_label(i)
# plt.subplots_adjust(left=0.05, right=2, top=0.95, bottom=0.5, wspace=0.3)
plt.tight_layout()
```

# Observation-

1) **Movies released in Afrikaans, Bosnian, Xhosa, and French have a higher chance of achieving high IMDb ratings.**

2) **Movies in the English language have mixed IMDb ratings, making it difficult to predict their relationship with the ratings due to the high number of movies released in this langeuag.**

3) **Movies released in Italian tend to hahigher chanceihood of receiving lower IMDb ratings, while movies in Japanese tend to have the opposite trend, with a higher chance of high ratigs.**

4) **Movies in Arabic and Hindi languages are more likely to have lower IMDb ratings.**
